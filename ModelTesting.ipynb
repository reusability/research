```python
from sklearn.cluster import KMeans
```

```python
kmean=KMeans(n_clusters=3)
```

```python
data_2 = data.drop('maven_reuse', inplace=False, axis=1)
data_3 = data_2.drop('project', inplace=False, axis=1)
data_4 = data_3.drop('release', inplace=False, axis=1)
data_5 = data_4.drop('maven_release', inplace=False, axis=1)
data_5.dropna(inplace=True)
```

```python
kmean.fit(data_5)
```

```python
kmean.cluster_centers_
```

```python
kmean.labels_
```

```python
wcss = []
for i in range(1,20):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(data_5)
    wcss.append(kmeans.inertia_)
    print("Cluster", i, "Inertia", kmeans.inertia_)
plt.plot(range(1,20),wcss)
plt.title("The Elbow Curve")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS") ##WCSS stands for total within-cluster sum of square
plt.show()
```
